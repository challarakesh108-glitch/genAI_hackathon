```python
import pandas as pd

orders = pd.read_csv("/content/orders.csv")

orders.head()
```

```python
users = pd.read_json("/content/users.json")

users.head()
```

```python
import sqlite3

conn = sqlite3.connect(":memory:")

with open("/content/restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)

restaurants = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants.head()
```

```python
orders_users = orders.merge(
    users,
    how="left",
    on="user_id"
)
```

```python
final_df = orders_users.merge(
    restaurants,
    how="left",
    on="restaurant_id"
)
```

```python
import pandas as pd

df = pd.read_csv("final_orders_dataset.csv")
df['order_date'] = pd.to_datetime(df['order_date'])

orders_over_time = df.groupby(df['order_date'].dt.to_period("M")).size()
orders_over_time.plot(title="Orders Over Time")
```

```python
user_behavior = df.groupby('user_id').agg(
    total_orders=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean')
)
```

```python
city_revenue = df.groupby('city')['total_amount'].sum().sort_values(ascending=False)
```

```python
cuisine_orders = df.groupby('cuisine')['order_id'].count()
```

```python
membership_stats = df.groupby('membership').agg(
    total_orders=('order_id', 'count'),
    avg_order_value=('total_amount', 'mean'),
    total_revenue=('total_amount', 'sum')
)
```

```python
df['total_amount'].hist(bins=30)
```

```python
avg_order_by_cuisine = (
    df.groupby("cuisine")["total_amount"]
    .mean()
    .sort_values(ascending=False)
)

avg_order_by_cuisine
```

```python
users_above_1000 = (
    df.groupby("user_id")["total_amount"]
    .sum()
    .reset_index()
)

count_users = users_above_1000[users_above_1000["total_amount"] > 1000]["user_id"].nunique()

count_users
```

```python
rating_bins = [3.0, 3.5, 4.0, 4.5, 5.0]
rating_labels = ["3.0-3.5", "3.6-4.0", "4.1-4.5", "4.6-5.0"]

df["rating_range"] = pd.cut(
    df["rating"],
    bins=rating_bins,
    labels=rating_labels,
    include_lowest=True
)
```

```python
revenue_by_rating = (
    df.groupby("rating_range")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

revenue_by_rating
```

```python
gold_df = df[df["membership"] == "Gold"]
```

```python
avg_order_by_city = gold_df.groupby("city")["total_amount"].mean().sort_values(ascending=False)
avg_order_by_city
```

```python
restaurants_per_cuisine = (
    df.groupby("cuisine")["restaurant_id"]
    .nunique()
    .reset_index(name="distinct_restaurants")
)

# Step 2: Total revenue per cuisine
revenue_per_cuisine = (
    df.groupby("cuisine")["total_amount"]
    .sum()
    .reset_index(name="total_revenue")
)
summary = restaurants_per_cuisine.merge(revenue_per_cuisine, on="cuisine")

summary
```

```python
total_orders = len(df)
gold_orders = len(df[df["membership"] == "Gold"])

percentage_gold = (gold_orders / total_orders) * 100
round(percentage_gold)
```

```python
restaurant_orders = df.groupby("restaurant_name_x")["order_id"].count().reset_index(name="total_orders")

# Step 2: Calculate average order value per restaurant
restaurant_avg = df.groupby("restaurant_name_x")["total_amount"].mean().reset_index(name="avg_order_value")

# Merge them
restaurant_summary = restaurant_orders.merge(restaurant_avg, on="restaurant_name_x")

# Step 3: Filter restaurants with < 20 total orders
filtered = restaurant_summary[restaurant_summary["total_orders"] < 20]

# Step 4: Find highest average order value among filtered
top_restaurant = filtered.sort_values("avg_order_value", ascending=False).head(1)

top_restaurant
```

```python
# Calculate total orders and average order value per restaurant
summary = df.groupby("restaurant_name_x").agg(
    total_orders=("order_id", "count"),
    avg_order_value=("total_amount", "mean")
).reset_index()

# Filter < 20 orders
filtered = summary[summary["total_orders"] < 20]

# Find highest average order value
top = filtered.sort_values("avg_order_value", ascending=False).head(1)

top
```

```python
df['order_date'] = pd.to_datetime(df['total_amount'])

# Extract quarter
df['quarter'] = df['total_amount'].dt.quarter

# Sum revenue by quarter
revenue_by_quarter = df.groupby('quarter')['total_amount'].sum().sort_values(ascending=False)

revenue_by_quarter
```

```python
options = [
    "Grand Cafe Punjabi",
    "Grand Restaurant South Indian",
    "Ruchi Mess Multicuisine",
    "Ruchi Foods Chinese"
]

summary = df[df["restaurant_name_x"].isin(options)].groupby("restaurant_name_x").agg(
    total_orders=("order_id", "count"),
    avg_order_value=("total_amount", "mean")
).reset_index()

filtered = summary[summary["total_orders"] < 20]
top = filtered.sort_values("avg_order_value", ascending=False).head(1)

top
```

```python
gold_df = df[df["membership"] == "Gold"]

avg_gold = gold_df["total_amount"].mean()

round(avg_gold, 2)
```

```python
high_rating_orders = df[df["rating"] >= 4.5]

# Count the orders
count_orders = len(high_rating_orders)

count_orders
```

```python
gold_df = df[df["membership"] == "Gold"]

# Step 1: Find the top revenue city among Gold members
city_revenue_gold = gold_df.groupby("city")["total_amount"].sum().reset_index()
top_city = city_revenue_gold.sort_values("total_amount", ascending=False).iloc[0]["city"]

# Step 2: Count how many orders were placed in that top city (Gold members only)
top_city_order_count = gold_df[gold_df["city"] == top_city].shape[0]

top_city, top_city_order_count
```
